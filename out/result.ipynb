```python
import json
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
with open("history.json", "r") as f:
    history = json.load(f)
```

```python
train_data = np.array(history["train"])
dev_data = np.array(history["dev"])
```

```python
plt.figure(figsize=(20, 10))
plt.subplot(231)
plt.plot(train_data[:, 0])
plt.plot(dev_data[:, 0])
plt.legend(["train-identify-P", "dev-identify-P"])

plt.subplot(232)
plt.plot(train_data[:, 1])
plt.plot(dev_data[:, 1])
plt.legend(["train-identify-R", "dev-identify-R"])

plt.subplot(233)
plt.plot(train_data[:, 2])
plt.plot(dev_data[:, 2])
plt.legend(["train-identify-F1", "dev-identify-F1"])

plt.subplot(234)
plt.plot(train_data[:, 3])
plt.plot(dev_data[:, 3])
plt.legend(["train-classify-P", "dev-classify-P"])

plt.subplot(235)
plt.plot(train_data[:, 4])
plt.plot(dev_data[:, 4])
plt.legend(["train-classify-R", "dev-classify-R"])

plt.subplot(236)
plt.plot(train_data[:, 5])
plt.plot(dev_data[:, 5])
plt.legend(["train-classify-F1", "dev-classify-F1"])
# plt.plot(history["train"])
```
